### Computation_Graph

<h2 style='color:red'> Limit : </h2>
    
Calculus'te limit, bir fonksiyonun bağımsız değişkeninin belirli bir değere yaklaştığında, fonksiyonun neye yaklaştığını ifade eden kavramdır. Özetle bir fonksiyondaki x değerleri belirli bir sayıya yaklaşırken fonksiyon grafiğinin nasıl davranacağı gözlemlenir.

"Limit ne işimize yarayacak" sorusu gelebilir. Doğrudan hiçbir işimize yaramayacak desek yanılmış olmayız. 

    Türev, bir niceliğin başka bir niceliğe göre değişimidir. Örneğin yolun zamana bağlı değişimi olan hız veya hızın zamana bağlı değişimi olan ivme türev kavramı ile tanımlanır. Limit aslında kavramsal olarak ortaya çıkmıştır ve Türevsel 
    işlemlerin yani değişimin olduğu her yerde bizler ile beraber olacaktır.
    
<b>Türev</b> için yardımcı elemandır diyebiliriz.

<h2 style='color:red'> Türev : </h2>

Türev, bir şeyin bir diğer şeye göre değişim miktarıdır. Yani türev, "değişim"i ölçmek için kullanılır. Örneğin, fizikte bir nesnenin hareketini inceleyelim. Nesnenin konumunu zamanla değiştiren bir fonksiyon olsun. Bu durumda fonksiyonun türevi, nesnenin hızını temsil eder. Çünkü Türev <b>"Değişim"</b> miktarı idi.

Türevin 2 farklı limit tanımı vardır, ikisini de türev hesaplamalarında kullanabiliriz:

<b>Türevin 1. Limit Tanımı : </b>

```python
"""Bir fonksiyonun x değişkenindeki değişimi sonucunda fonksiyon değerinde 
oluşan ANLIK DEĞİŞİM ORANINA, fonksiyonun x değişkenine göre türevi denir."""

display.Image("pngs/Computation Graph_1.png")

# L = f'(a) >>> Bu limit değerine fonksiyonun bu noktadaki türevi denir.
```

Limit tanımına bağlı olarak şöyle bir yorum yapabiliriz değil mi ? 

* x ----> a ya doğru çok az miktar yaklaşmış olsun. Bu durumda fonksiyonun f(x) değeri ile f(a) değeri arasındaki fark fonksiyonun anlık değişimidir. 

<b>Türev ne idi ?</b>

Fonksiyon değerinde oluşan <b style='color:red'> ANLIK DEĞİŞİM ORANI idi. </b> Bu mantık ile, f(x) - f(a) farkının (x - a) ile bölümü fonksiyonun x değişkenine göre türevini vermiş oldu.

```python
# ÖRNEK

# Bir fonksiyonun belirli bir x noktasındaki TÜREV DEĞERİ fonksiyonun grafiğine o noktada çizilen teğetin eğimine eşittir.

"Aşağıdaki örnekte x = 3 noktasındaki türev değeri, fonksiyonun bağımsız(x) değişkeninin çok küçük bir değişimi karşısında \
bağımlı değişkenin(f(x)) anlık değişim oranını temsil eder."

display.Image("pngs/Computation Graph_2.png")
```

* Tanımda h 0'a yaklaşacak şekilde <b>h = x - a</b> dönüşümü yaparsak türevin daha sık kullanılan ikinci limit tanımını elde ederiz.
