<h2 style='color:red'> Limit : </h2>
    
Calculus'te limit, bir fonksiyonun bağımsız değişkeninin belirli bir değere yaklaştığında, fonksiyonun neye yaklaştığını ifade eden kavramdır. Özetle bir fonksiyondaki x değerleri belirli bir sayıya yaklaşırken fonksiyon grafiğinin nasıl davranacağı gözlemlenir.

"Limit ne işimize yarayacak" sorusu gelebilir. Doğrudan hiçbir işimize yaramayacak desek yanılmış olmayız. 

    Türev, bir niceliğin başka bir niceliğe göre değişimidir. Örneğin yolun zamana bağlı değişimi olan hız veya hızın zamana bağlı değişimi olan ivme türev kavramı ile tanımlanır. Limit aslında kavramsal olarak ortaya çıkmıştır ve Türevsel 
    işlemlerin yani değişimin olduğu her yerde bizler ile beraber olacaktır.
    
<b>Türev</b> için yardımcı elemandır diyebiliriz.

<h2 style='color:red'> Türev : </h2>

Türev, bir şeyin bir diğer şeye göre değişim miktarıdır. Yani türev, "değişim"i ölçmek için kullanılır. Örneğin, fizikte bir nesnenin hareketini inceleyelim. Nesnenin konumunu zamanla değiştiren bir fonksiyon olsun. Bu durumda fonksiyonun türevi, nesnenin hızını temsil eder. Çünkü Türev <b>"Değişim"</b> miktarı idi.

Türevin 2 farklı limit tanımı vardır, ikisini de türev hesaplamalarında kullanabiliriz:

Optimizasyon Problemleri: Türev ve gradyan, optimizasyon problemlerinde büyük öneme sahiptir. Bir fonksiyonun en küçük veya en büyük değerini bulmak için türev ve gradyan kullanarak fonksiyonun minimum veya maksimum noktalarını tespit edebiliriz. Bu, mühendislik, ekonomi, istatistik gibi birçok alanda kullanılan optimizasyon problemlerini çözmekte yardımcı olur.

Eğim Tabanlı Algoritmalar: Makine öğrenmesi ve yapay zeka alanlarında, gradyan tabanlı optimizasyon algoritmaları kullanılır. Bu algoritmalar, bir fonksiyonun gradyanını kullanarak bir modelin parametrelerini ayarlamayı ve hedefi en iyi şekilde optimize etmeyi hedefler. Örneğin, gradient iniş algoritması, bir hata fonksiyonunu minimize etmek için gradyanı kullanır ve modelin en iyi parametre değerlerini bulmaya çalışır.

Evet, türevin 0 olduğu noktalarda fonksiyonun eğimi de 0'dır. Türev, bir fonksiyonun anlık değişimini temsil ettiğinden, türevin sıfır olduğu bir noktada fonksiyonun anlık değişimi de sıfır olur. Bu da, fonksiyonun eğiminin 0 olduğunu gösterir.

Geometrik olarak düşünürsek, bir fonksiyonun grafiği üzerindeki bir noktada türevin 0 olduğu noktalar, fonksiyonun düzlük (horizontal) veya tepe (peak) noktalarını gösterir. Bu noktalarda fonksiyonun eğimi 0 olduğu için, fonksiyon ne artar ne de azalır. Eğer nokta lokal bir minimum veya maksimum noktasıysa, türevin sıfır olduğu noktada eğim geçici olarak 0 olur ve eğim fonksiyonun hangi yönde değiştiğine bağlı olarak artar veya azalır.

Ancak, türevin 0 olduğu bir noktanın her zaman bir minimum veya maksimum noktası olmadığını unutmamak önemlidir. Bu noktalar ayrıca düzlük veya tepe noktaları olabilir, ancak gerçek bir minimum veya maksimum nokta olup olmadığını belirlemek için daha fazla analiz yapmak gerekir.
