<h2 style='color:red'> Limit : </h2>
    
Calculus'te limit, bir fonksiyonun bağımsız değişkeninin belirli bir değere yaklaştığında, fonksiyonun neye yaklaştığını ifade eden kavramdır. Özetle bir fonksiyondaki x değerleri belirli bir sayıya yaklaşırken fonksiyon grafiğinin nasıl davranacağı gözlemlenir.

"Limit ne işimize yarayacak" sorusu gelebilir. Doğrudan hiçbir işimize yaramayacak desek yanılmış olmayız. 

    Türev, bir niceliğin başka bir niceliğe göre değişimidir. Örneğin yolun zamana bağlı değişimi olan hız veya hızın zamana bağlı değişimi olan ivme türev kavramı ile tanımlanır. Limit aslında kavramsal olarak ortaya çıkmıştır ve Türevsel

<h2 style='color:red'> Türev : </h2>

Türev, bir şeyin bir diğer şeye göre değişim miktarıdır. Yani türev, "değişim"i ölçmek için kullanılır. Örneğin, fizikte bir nesnenin hareketini inceleyelim. Nesnenin konumunu zamanla değiştiren bir fonksiyon olsun. Bu durumda fonksiyonun türevi, nesnenin hızını temsil eder. Çünkü Türev <b>"Değişim"</b> miktarı idi.

<h4> c.1 ) Gradyan Vektörü : </h4>

<b>Örnek; Üç Değişkenli Fonksiyon:</b>
    
Fonksiyon: f(x, y, z) = 2x^2 + 3y - z

Bu fonksiyonun gradient vektörünü (∇f) bulalım:
    
∇f = (∂f/∂x, ∂f/∂y, ∂f/∂z)

*    ∂f/∂x = 4x (x'ye göre türev alındığında);

*    ∂f/∂y = 3  (y'ye göre türev alındığında);

*    ∂f/∂z = -1 (z'ye göre türev alındığında)

<b>Sonuç olarak, gradient vektörü (∇f) = (4x, 3, -1) olur.</b>

-----------------

<b style='color:red'> Gradyan vektörü (∇f) = (4x, 3, -1) kısmi türevleri nasıl kullanılacak :  </b>

* Kısmi türevlerin bileşkesi olan gradyanın yönü, değişkenlerin hangi yönde değişmeleri halinde fonksiyonun artacağını gösterir. O halde gradyan vektörü yönünde ilerlenirse maksimuma, gradyanın yönünün tersi takip edilirse minimuma yaklaşılır.

<b style='color:red'>Literatürdeki tanımları ile Gradient vektörü: </b>

    "The gradient vector gives us the direction with the steepest ascent."

    "Gradient is vector that points in the direction of greatest increase of a function."

<b>Steepest descent ve Steepest ascent kavramları: </b>


* Gradyan vektörünün tersi yönde ilerlenir ise "steepest ascent" in tam tersi yönde ilerlemiş oluruz ki buna "steepest descent" denir.  Gradient Descent' in temeli de buradadır; gradyanın tersi yönde iniş yapılır.

* Gradyan iniş(gradient descent), bir fonksiyon üzerinde rastgele bir noktadan başlayıp bu noktanın koordinatlarını, kısmi türevlerin tersi (yani gradyan vektörünün tersi) yönünde değiştirerek küçük adımlarla minimum noktaya yaklaşma yöntemidir. 

```python
# "Steepest Descent" in yönü "Gradient" yönünün tersidir.

display.Image("pngs/Basic_Concepts_13.png")
```

<b>ÖRNEK 1 : Diyelim ki bir fonksiyonumuz f(x) = x^2 olsun ve bu fonksiyonun minimum noktasını bulmak istiyoruz.</b> 

Gördüğümüz gibi, fonksiyonun minimum noktası x = 0'da yer almaktadır.

Gradient Descent algoritmasında, başlangıç noktası olarak x = 3 seçelim. Şimdi adımları takip edelim:

    Başlangıç noktası: x = 3

    Gradient hesapla: ∇f(x) = 2x = 2 * 3 = 6

    Adım boyutunu(learning rate) belirle: Örneğin adım boyutunu 0.1 olarak seçelim.

    Yeni noktayı hesapla: x_new = x - learning_rate * gradient = 3 - 0.1 * 6 = 2.4

    Kontrol et: Fonksiyonun değeri azaldı mı? f(2.4) = (2.4)^2 = 5.76 < f(3) = (3)^2 = 9

    Devam et: Yeni nokta olan x = 2.4 üzerinde işlemleri tekrarla.

    Bu adımları tekrar ettiğimizde, gradientin tersi yönde ilerleyerek fonksiyonun değeri azaltacağız. 
    Yani x değeri her adımda azalacak ve fonksiyon minimum noktasına yaklaşacaktır.

```python
display.Image("pngs/Basic_Concepts_14.png")
```

#####  Yukarıda "Kontrol et:" adımı var. Bu adım neden var ? 

Bu kontrol adımı, algoritmanın doğru yönde ilerleyerek fonksiyonun minimum noktasına yaklaşmasını sağlar. Fonksiyon değerinin azalması, <b>gradientin tersi yönde doğru</b> ilerlediğimizi gösterir ve algoritmanın doğru yönde güncellemeler yapmasını sağlar.

Gradient Ascent: Fonksiyonun maksimum noktasını bulmak için kullanılır. Maksimum noktaya doğru ilerlerken, adımlarını gradientin yönünde atarız. Bu nedenle, her adımda yeni x değerini hesaplarken toplama işlemi kullanılır: x = x + α * ∂f/∂x, y = y + α * ∂f/∂y

Gradient Descent: Fonksiyonun minimum noktasını bulmak için kullanılır. Minimum noktaya doğru ilerlerken, adımlarını gradientin tersi yönde atarız. Bu nedenle, her adımda yeni x değerini hesaplarken çıkarma işlemi kullanılır: x = x - α * ∂f/∂x, y = y - α * ∂f/∂y
